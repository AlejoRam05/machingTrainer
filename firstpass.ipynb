```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BreastCancer.csv')
```

```python
# Habilita la visualización de gráficos en línea dentro del notebook.
%matplotlib inline

# Establece el estilo visual de los gráficos de Seaborn a 'whitegrid'.
# Esto agrega una cuadrícula gris clara al fondo de los gráficos.
sns.set_style(style='darkgrid')

# Define el contexto de los gráficos de Seaborn como 'notebook'.
# Esto ajusta el tamaño de los elementos del gráfico (texto, líneas, etc.)
# para una mejor visualización en un notebook.
sns.set_context(context='talk')

# Establece el tamaño por defecto de las figuras de Matplotlib a 11 pulgadas de ancho
# y 9.4 pulgadas de alto.
plt.rcParams['figure.figsize'] = (11, 9.4)
```

```python
g=(
    df
    .isnull()
    .melt(value_name='missing')
    .pipe(
        lambda df:(
            sns.displot(
                data=df,
                y='variable',
                hue='missing',
                multiple='fill',
                aspect=2,
                stat='percent'

            )
        )
    )

)

plt.show()
```
